#### 2:

## Koodi
Nopea esimerkki datan kartoittavasta visualisoinnista

```python
## haetaan kirjastot
import pandas as pd
import matplotlib.pyplot as plt

## Luetaan tarvittava data
data = pd.read_csv("listings.csv")

## Luodaan matplotlibin scatter plot
## Yritetään tutkia hinnan ja arivioiden määrän suhdetta
plt.plot(data['number_of_reviews'])
plt.plot(data['price'])
## Asetetaan otsikko
plt.title("Scatter plot of Airbnb listing data")
## Asetetaan x- ja y akselin tunnisteet
plt.xlabel("Price")
plt.ylabel("Reviews")
plt.show()
```

```python
## Kuten huomataan, edellinen on hyvä esimerkki huonosta visualisoinnista
## Suhde muuttujien välillä ei selkene datasta yhtään, vaikka on toki
## erittäin todennäköistä, että suhdetta ei edes ole.
## Luodaan vaihtoehtoinen pylväsdiagrammi, ja tarkastellaan auttaako se
## visualisoinnissa
plt.bar(data['price'],data['number_of_reviews'])
plt.title("A bar chart")
plt.xlabel("Price")
plt.ylabel("Number of reviews")
plt.show()
```
